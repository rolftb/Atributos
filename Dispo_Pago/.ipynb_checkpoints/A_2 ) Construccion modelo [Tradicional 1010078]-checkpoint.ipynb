````python
# =============================================================================
# Ajustar la data
# =============================================================================
def Ajuste_Clientes_RFM(data):
    start_time = time.time()
    df=data.copy()

# =============================================================================
# Proceso iterativo Pasar a numero
#------------------------------------------------------------------------------

#----------------------------------------------------------------------
# Float
    for i in [
            #RFM
            'Recency Actual'
            ,'Frecuency Actual del Cliente'
            ,'Ingreso de Venta Promedio'
              
            #Val Fact
            ,'Kilos Venta KG Promedio'
            ,'Precio Promedio'
            
            # Val historicos (ind del producto)
            ,'Año Creación Promedio'
            ,'Frecuencia Historica Promedio'
             ]:
        df[i]=df[i].astype('float64')
        #.astype('int64')
        
        
#------------------------------------------------------------------------ 
    # Int
    for i in [
                #RFM
                'Recency Actual'
#                 ,'Frecuency Actual del Cliente'
#                 ,'Ingreso de Venta Promedio'

                #Val Fact
#                 ,'Kilos Venta KG Promedio'
#                 ,'Precio Promedio'

                # Val historicos (ind del producto)
                ,'Año Creación Promedio'
                ,'Frecuencia Historica Promedio'
                ]:
        df[i]=df[i].astype('int64')
        


#==========================================================================================
# Ajuste de las fechas
#-----------------------------------------------------------------------------------------
    for date in ['Fecha ultima Compra','Fecha Creacion']:
        df[date] = df[date].apply(lambda x: 
                                            datetime.datetime.strptime(str(x),
                                                                           '%Y-%m-%d').date())
    
    print(df.dtypes)
    end_time = time.time()
    time_convert(end_time - start_time)
    return df
````

# Regresión Ridge
### Suavisa el error de la varianza
En este sentido (y asumiendo un modelo lineal), cuanto mayor
es $k$ más grande es el sesgo pero menor es la variancia del
estimador, y la determinaci´on final implica un compromiso
entre ambos términos (Izenman, 2008).
## Peso(4 dimensiones) + Semana
### Evaluación de los coeficientes 

\begin{equation}
    P(Q,W,A)=\beta_1Q+\sum_{i\in W}\beta_iw_i+\beta_0
\end{equation}

* $P$ Precio Estimado
* $Q$ Peso neto KG
* $w_i$ Semana $i$ del conjunto de semanas $W$

### Grafico Dispersión del precio con predicción
No posee limite para el lote de compra, por lo cual se consideran compras sobre 3 toneladas, se observa una grafica, con una caida abrupta al sobrepasar las 3 toneladas. Sucesivamente se tiene la misma grafica, sin contemplar este escenario.

#### Predicción 

##### Predicción

###### DAtos 

````python
#===================================
#Ajuste de los datos
#-----------------------------------
y=Data[['Precio']].copy()
X=Data[['Kilos Venta KG','Año']].copy()
#X['Semana']=X['Semana'].astype('category')
X['Año']=X['Año'].astype('int64')
X['Año']=X['Año']-2019
X['Kilos Venta KG 2']=Data[['Kilos Venta KG']]**2
X['Kilos Venta KG 3']=Data[['Kilos Venta KG']]**3
X['Kilos Venta KG 4']=Data[['Kilos Venta KG']]**4
#X.dtypes

X=pd.get_dummies(X, drop_first=False)
#X.columns
````

````python

#===================================
# Entrenamiento
#-----------------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train, y_train)


#===================================
# Puntaje Prueba, Test
#-----------------------------------

print("Training set score: {:.2f}".format(ridge.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge.score(X_test, y_test)))
#print("Las columnas del modelo:",(X.columns))


#===================================
# Formula
#-----------------------------------
print("La variación del precio con")
print("respecto al tamaño de lote es la siguiente función")
print(" como X Kilos Venta KG: ",(ridge.coef_[0][0]),'X + ',
      (ridge.coef_[0][2]),'X^2 + ',(ridge.coef_[0][3]),'X^3'
      ,'\n')


#===================================
# ejemplo de calcular una caja
#-----------------------------------
print('Lo que se traduce que por aumentar un lote en 22 kilogramos')
# No se trata de aumentar, porque depende del valor que se está aumentado, es decir de x_1
print('el Precio CLP deberia ser en {:.2f}'
      .format(ridge.coef_[0][0]*22+ridge.coef_[0][2]*22**2+
              ridge.coef_[0][3]*22**3+ridge.intercept_[0]),'CLP \n',
              'para la primera semana de enero')
print('\nEl intercepto es {:.2f}'.format((ridge.intercept_[0])))
````

## OLS 


OLS posee en la libreria el [p-value](https://www.cluzters.ai/forums/topic/395/find-p-value-significance-in-scikit-learn-linear-regression?c=1597)

Entonces se provará la utilidad de las variables independientes

Tenga en cuenta que es posible obtener un R-cuadrado negativo para ecuaciones que no contienen un término constante. Debido a que R-cuadrado se define como la proporción de la varianza explicada por el ajuste, si el ajuste es realmente peor que simplemente ajustar una línea horizontal, entonces R-cuadrado es negativo. 
En este caso, R-cuadrado no puede interpretarse como el cuadrado de una correlación. Tales situaciones indican que se debe agregar un término constante al modelo.




[P-value interpretación](https://blog.minitab.com/es/como-interpretar-los-resultados-del-analisis-de-regresion-valores-p-y-coeficientes)

[p-value significacia](https://scielo.isciii.es/scielo.php?script=sci_arttext&pid=S1139-76322017000500014)

### Simple

\begin{equation}
P(Q,W,A)=\beta_1Q +\sum_{i\in W}\beta_iw_i +\beta_a A +\beta_0
\end{equation}



- $P:$ Precio Estimado
- $Q:$ Peso neto Kg.(_Variable discreta_)
- $A:$ Año el cual se ejecuta la venta.(_Variable discreta_)
- $w_i$ Semana $i$ del conjunto de semanas $W$

###### Función 

````python
import re
def Col_Dummy(txt,Col_df,ind):
    """
    Extrae la columna original ( Col_df) del dummy,
    Si no es dummy, cuando ind = 1, entrega un 0. Else, entrega la categoria del dummy
    """
    
    for c in X_Columnas:
        if(re.search('^'+c, txt) is not None):
            if(re.search('^'+c+'_', txt) is not None):
                post=re.search('^'+c+'_', txt).span()[1]
                return(c,txt[post:])[ind]
            else:
                return(c,0)[ind]
    return(0,0)[ind]
df_Var_Modelo['Atributo']=df_Var_Modelo.Variable.apply(lambda x: Col_Dummy(x,X_Columnas,0) )
df_Var_Modelo['Val Categoria']=df_Var_Modelo.Variable.apply(lambda x: Col_Dummy(x,X_Columnas,1) )

#=========================================
# Ajuste de Val Categoria a numero entero
#-----------------------------------------

df_Var_Modelo['Val Categoria']=df_Var_Modelo['Val Categoria'].astype('float64')

df_Var_Modelo['Val Categoria'] =df_Var_Modelo['Val Categoria'].astype('int64')

#Ordenar en base al numero de la semana
df_Var_Modelo=df_Var_Modelo.sort_values(by=['Val Categoria']).reset_index()

#eliminar la columna index
df_Var_Modelo=df_Var_Modelo.drop(labels='index',axis=1)

df_Var_Modelo.head()
````

#### Interpretación Coeficiente 

Interpretación $\beta$ en para $w_i$
\begin{equation}
\dfrac{y_2}{y_1} =e^{\beta(x_2-x_1)}\\
    \%\triangle Y = e^{\beta(\triangle X)}
\end{equation}

La ecuación **Log-Nivel**, da ceunta que ante un cambio unitario de $X$ representa una alteración porcentual en la variable dependiente $Y$

En este caso se toma de precio base el precio de la primera semana del año. Esto sujeto a la misma relación precio/peso neto, del producto.

````python
df_semana_OLS.columns
````

````python
# df_semana_OLS[['Val Categoria','Coef.']]
Var_trad={'Variación porcentual del precio':'$ \% \triangle Y$'}
df_semana_OLS['$\%\triangle Y$']=df_semana_OLS['Coef.'].apply(lambda x: np.e**x-1)
df_semana_OLS[['Semana','$\%\triangle Y$',]]

````
